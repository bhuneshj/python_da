### Plot1

#### A plot showing the fraction of total investments (globally) in venture, seed, and private equity, and the average amount of investment in each funding type. This chart should make it clear that a certain funding type (FT) is best suited for Spark Funds.

```python
# Calulcating Sum, Count and Mean for each funt types
df = master_frame.groupby('funding_round_type')['raised_amount_usd'].agg(['sum','count','mean'])
# Getting fund types whcih are required
top_four_funds_df = (df.loc[df.index.isin(["venture","angel","seed","private_equity"])]).copy()
# Calculating Investment count fraction gloabally for each fund types
top_four_funds_df["Investment_Count%"] = top_four_funds_df["count"]/sum(top_four_funds_df["count"])*100
top_four_funds_df["Raised_Amount_USD"] = top_four_funds_df["sum"]
top_four_funds_df["Average_Amount_Raised"] = top_four_funds_df["mean"]
top_four_funds_df["Number_of_Investments"] = top_four_funds_df["count"]
top_four_funds_df
```
